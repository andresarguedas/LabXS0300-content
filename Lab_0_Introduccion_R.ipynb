## Tipos de variables y datos en R

### Tipos de variables

Hasta ahora, hemos visto como trabajar con nmeros en R, pero realmente hay muchos distintas formas en que podemos manipular nuestros datos en R. Las principales formas se presentan a continuacin:

- **Numricos:** Representa un nmero, puede ser natural, entero, real o hasta imaginario. Son el tipo de variable que ms vamos a usar, y por lo general la mayor parte de las funciones estn diseadas para nmeros, pero no son el nico tipo de variable que usaremos.
- **Caracteres** o **strings:** Representan texto en general, sin ningn orden, y pueden aceptar casi cualquier tipo de informacin como texto. Es una de las formas ms verstiles de guardar informacin en R, pero tiene el inconveniente que, al ser tan libre, es poco estructurado y se puede volver difcil trabajar con estas.
- **Factores:** Son un tipo especial de variable de texto o numrica, donde la variable solamente puede tomar ciertas valores. Por ejemplo, si tenemos una cocina que solamente tiene tres niveles de calor: Alto, Medio y Bajo, podramos definir una variable de factor, donde los *niveles* de dicha variable sean los posibles calores: `Alto`, `Medio` y `Bajo`. Cuando trabajemos con datos reales en el prximo laboratorio, veremos cmo funcionan los factores con mayor profundidad.
- **Lgicos:** Se utilizan para denominar variables que son verdaderas (`TRUE` o `T`) o falsas (`FALSE` o `F`). Por lo general, se utilizan para determinar, dentro de las funciones, si ciertas condiciones se estn cumpliendo, para poder aplicar los procedimientos adecuados.

Adems de los mencionados anteriormente, R tambin soporta distintos tipos de variables como fechas, imagenes, sitios web, coordenadas geogrficas, mapas y se pueden definir ms tipos mediante distintos paquetes. Con respecto a la forma en la cual podemos almacenar dichos datos, las dos formas ms comnes para hacerlo son mediante *vectores* o *data frames*.

### Vectores

En primer lugar, los vectores son una serie de elementos de un cierto tipo (numricos, caracteres, fechas, etc.) que se guardan como una sola variable. Por ejemplo, supongamos que los de tres estudiantes son 70, 80 y 50 kg, respectivamente. En este caso, podemos guardar esta informacin como tres variables distintas (`peso1`, `peso2` y `peso3`) o podemos guardar los tres pesos en una sola variable (`peso`), que sera un vector. Primero guardemos los pesos como tres variables distintas:

````r
peso1 <- 70
peso1
peso2 <- 80
peso2
peso3 <- 50
peso3
````

En este caso, tanto `peso1`, `peso2` como `peso3` son variables numricas que contiene solamente un valor, el peso de cada estudiante. Ahora, si quisieramos hacer un vector con los pesos de los 3 estudiantes, podemos usar la funcin `c()` para crear un vector, donde los argumentos de dicha funcin son los valores de los pesos, como se muestra a continacuin:

````r
peso <- c(70, 80, 50)
peso
````

Como podemos ver, el resultado de `peso` es distinto al de las tres variables que definimos anteriormente, dado que obtuvimos los tres valores en una sola lnea. Esto se debe a que la variable `peso` tiene tres "posiciones" distintas, que son los "espacios" correspondientes a cada uno de los pesos de los estudiantes. Por lo anterior, el peso del primer estudiante ocupa la primer posicin, el peso del segundo estudiante la segunda posicin y el peso del tercer estudiante la tercera posicin. Si queremos obtener solamente algunos de los pesos de los estudiantes, podemos agregar `[]` al final de `peso` y especificar la posicin, o las posiciones, que queramos obtener. Veamos el siguiente ejemplo:

````r
# Peso del primer estudiante
peso[1]
# Peso del segundo estudiante
peso[2]
# Peso del tercer estudiante
peso[3]
# Peso del segundo y tercer estudiante
peso[2:3] # En este caso 2:3 devuelve los nmeros del 2 al 3
# Peso del primer y tercer estudiante
peso[c(1, 3)] # Podemos definir un vector dentro de parntesis cuadrados
````

En la celda anterior, note que hay cierto texto en verde despus de un signo numeral (`#`), esto se conoce como un comentario y cualquier cosa en una lnea que se escriba despus de un `#` no ser ejecutado. ```

sin tomar en cuenta el texto en verde. Con respecto a los resultados obtenidos, cmo se puede ver, si queremos guardar informacin de varios sujetos para una misma variable, es ms eficiente utilizar un vector que guardar cada dato como una variable distinta.

### Data frames

Un data frame es la forma de almacenar datos ms usado en R y una de las principales razones por las cuales este lenguaje es uno de los ms usados para anlisis estadsticos a lo largo de distintas disciplinas. De forma general, un data frame es una agrupacin de vectores del mismo tamao, por ejemplo, si ahora tenemos la siguiente informacin de los estudiantes:

|Estudiante|Peso|Altura|Edad|
|:--------:|:--:|:----:|:--:|
|1         |70  |180   |20  |
|2         |80  |160   |18  |
|3         |50  |150   |22  |

podramos definir tres vectores distintos, uno para cada una de las variables deseadas (`peso`, `altura` y `edad`), o podemos crear un data frame que contenga toda la informacin de los estudiantes (`estudiantes`). En el caso de los tres vectores, uno para cada variable, haramos lo siguiente:

````r
peso <- c(70, 80, 50)
peso
altura <- c(180, 160, 150)
altura
edad <- c(20, 18, 22)
edad
````

Como podemos ver, tenemos tres vectores distintos, de tamao tres, donde el primer valor de cada vector corresponde a informacin del primer estudiante y as con todos los dems estudiantes. Si lograramos juntar todos estos vectores en un solo objeto, podramos tener las variables como columnas y los estudiantes como filas, y esto es lo que hace un data frame. Para crear un data frame, usamos la funcin `data.frame` y agregamos como argumentos a dicha funcin los nombres de las variables, junto con los datos para dicha columna, es decir, los vectores que creamos anteriormente. El resultado de la celda anterior es distinto al obtenido mediante los tres vectores, ya que aparece como una table, donde nos dice el nombre de las columnas (que son las variables que nosotros definimos como argumentos) y cada fila de esa tabla representa a un estudiante. Adems, si presionan sobre los nombres de las variables, pueden ordenar las filas segn dicha variable, de mayor a menor o vice versa. Por ltimo, debajo del nombre de las variables est viene un `<dbl>`, esto nos indica de que tipo es la variable presentada y, en este caso, eso se le conoce como un *double*, que es simplemente una variable numrica.

Ahora, los data frames tienen ciertas ventajas y capacidades nicas que son tiles de conocer. La primera de estas es que, para obtener solamente una de las variables, o columnas, del data frame, la podemos llamar agregando un `$` despus del nombre del data frame y agregando el nombre de la variable que nos interesa. Por ejemplo, si queremos obtener las edades de los estudiantes, podemos hacer lo siguiente:

````r
estudiantes$Edad
````

En este caso, lo que no est devolviendo es un vector con las edades de los estudiantes, que es igual al objeto `edad` que definimos anteriormente, por lo que podemos aplicar lo que ya conocemos sobre vectores a este objeto. Ahora, qu pasa si queremos solamente la informacin del primer estudiante? Esto lo podemos lograr utilizando nuevamente los `[]`:

````r
# Informacin del primer estudiante
estudiantes[1, ]
# Informacin del segundo estudiante
estudiantes[2, ]
# Informacin del tercer estudiante
estudiantes[3, ]
# Informacin del segundo y tercer estudiante
estudiantes[2:3, ]
# Informacin del primer y tercer estudiante
estudiantes[c(1, 3), ]
````

La forma de utilizar los `[]` es igual a como lo vimos para los vectores, solamente que ahora tenemos que tomar en cuenta que los data frames tienen filas y columnas, por lo que tenemos que escribir `[fila, columna]`. Si dejamos la fila o la columna vaca, como hicimos en los ejemplos anteriores, R toma por defecto todas las columnas o filas, respectivamente, pero tambin podramos hacer selecciones ms complejas de datos dependiendo de las filas y columnas que querramos obtener.

## Paquetes de R

Como hemos visto, R tiene una gran variedad de funciones tanto para hacer operaciones matemticas como para guardar datos y manipularlos, pero todava podemos agregar incluso ms funciones a R. Hay una gran cantidad de paquetes libres en internet, la mayora disponibles mediante CRAN (The **C**omprehensive **R** **A**rchive **N**etwork), un repositorio de paquetes de R, pero tambin hay paquetes creados por empresas o individuos para uso personal o comercial. Una lista de los paquetes disponibles en CRAN se puede consultar en [este enlace](https://cran.r-project.org/web/packages/available_packages_by_name.html).

Ya habiendo instalado un paquete, ocupamos cargarlo a R, dado que no tiene sentido que todos los paquetes estn corriendo en todo momento, sino simplemente cuando los ocupamos. ### El paquete `dplyr` y el operador `%>%`

El paquete que acabamos de cargar `dplyr` fue creado con el objetivo de mejorar la legibilidad de los cdigos producidos en R, sobre todo en lo referente a la manipulacin de conjuntos de datos y es parte de un gran conjunto de paquetes, denominado el [`tidyverse`](https://www.tidyverse.org), donde todos buscan mejorar la experiencia de trabajar con R. A lo largo del curso veremos varias funciones de `dplyr`, y de otros paquetes del `tidyverse`, pero para este laboratorio nos vamos a enfocar en la funcin `%>%` u operador **pipe**, como es conocido en ingls. Para agregar este operador en la consola de R, pueden presionar **Ctril + Shift + Enter** y automticamente agrega el operador a la consola.

Dicho operador se utiliza de la siguiente forma. Supongamos que tenemos un cierto objeto `x` y le queremos aplicar una funcin `f()`, entonces podramos escribir lo siguiente:

```{r}
f(y)
```

Usando el operador pipe, lo mismo se puede escribir de la siguiente forma:

```{r}
y %>% f()
```

Cul es la ventaja de utilizar un pipe? Tal vez no es tan aparente con una sola funcin, pero veamos un ejemplo ms complicado. Queremos aplicarle al objeto `y` la funcin `f()`, luego la funcin `g()`, posteriormente la funcin `h()` y, por ltimo, la funcin `k()`. Escribiendo esto, lo podemos hacer de la siguiente forma:

```{r}
k(h(g(f(y))))
```

O, usando el operador pipe, lo podemos escribir as:

```{r}
y %>% f() %>% g() %>% h() %>% k()
```

Como podemos ver, el comando usando pipes se entiende de forma ms sencilla como lo que queremos realizar: aplicarle a `y` las funciones `f()`, `g()`, `h()` y `k()`, respectivamente, mientras que, escribiendolo sin un pipe, se vuelve confuso lo que estamos haciendo. Ahora, si las funciones tuvieran otros argumentos, el primer cdigo se volvera ilegible, mientras que usando el pipe se seguira viendo legible y se entendera de mejor forma. El resultado de ambas formas de escribir el cdigo es igual, por lo que, como hemos visto, podemos utilizar el operador de asignacin `<-` para guardar un resultado de un comando usando pipes. Por lo tanto, si quisieramos calcular algo como $x = \sqrt{e^2}$, lo podramos calcular de la siguiente forma:

````r
x <- 2 %>% exp() %>% sqrt()
x
````

**Recuerde que ocupa correr el comando `library(dplyr)` antes de ejecutar un pipe, o sino le dar un error.**

Como podemos ver, estamos guardando en una variable, llamada `x`, el resultado de la operacin `2 %>% exp() %>% sqrt()`, que la podemos leer como que a 2 le estamos aplicando la funcin exponencial, $e^2$, y a esto le estamos sacando la raz cuadrada, $\sqrt{e^2}$, que es lo que queramos obtener. Este resultado sera igual a si corrieramos `x <- sqrt(exp(2))`, pero es ms fcil de entender mediante pipes.
